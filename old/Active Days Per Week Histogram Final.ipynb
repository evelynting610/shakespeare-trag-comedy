```python
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
pdropoffs = pd.read_csv("pinterest/datastory/fd_active_days_per_week.csv")
pactive = pd.read_csv("pinterest/datastory/normal_active_days_per_week.csv")
```

```python
pdropoffs = pdropoffs.groupby('expert').aggregate(np.median)
pactive = pactive.groupby('expert').aggregate(np.median)
```

```python
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),    
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]    
  
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.    
for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.)    
```

```python
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
plt.figure(figsize=(17, 10))    
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)    
  
# Ensure that the axis ticks only show up on the bottom and left of the plot.    
# Ticks on the right and top of the plot are generally unnecessary chartjunk.    
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()
ax.set_title('Days per Week Pinterest Experts are Active', fontsize= 20)
ax.set_ylabel('Percent of Experts', fontsize = 16)
ax.set_xlabel('Number of Days Active', fontsize = 16)

plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
data1 = pdropoffs['days_active']
data2 = pactive['days_active']
plt.hist(data1, 14, range=(0, 7), normed=False, weights=np.zeros_like(data1) + 100. / data1.size, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', label = 'dropoffs (n=18)', alpha=0.9, color = tableau20[0])
plt.hist(data2, 14, range=(0, 7), normed=False, weights=np.zeros_like(data2) + 100. / data2.size, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', label = 'nondropoffs (n=17)', alpha =0.6, color = tableau20[6])
plt.legend(loc='upper left')
```
