```python
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("airbnb/lastweek_responses_by_day.csv")
```

```python
df.plot()
```

```python
responses = df[['weekday', 'responses']].copy()
```

```python
responses[:20]
```

```python
weekday_counts = responses.groupby('weekday').aggregate(np.median)
weekday_counts.index = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
weekday_counts
```

```python
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
plt.figure(figsize=(15, 8))    
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)    
  
# Ensure that the axis ticks only show up on the bottom and left of the plot.    
# Ticks on the right and top of the plot are generally unnecessary chartjunk.    
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()
ax.set_title('Activity of Dropoffs their Last Week on Directly', fontsize= 20)
ax.set_ylabel('Median Number of Responses', fontsize = 16)
ax.set_xlabel('Days', fontsize = 16)

plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
weekday_counts.plot(kind = 'bar', ax=ax, lw=2.5, rot=45)
```

```python
df = pd.read_csv("airbnb/lastweek_responses_by_day.csv")
all_responses = df[['expert', 'weekday', 'responses']]
```

```python
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),    
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]    
  
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.    
for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.) 
```

```python
all_responses.loc[all_responses.weekday==1, ['weekday']]='Monday'
all_responses.loc[all_responses.weekday==2, ['weekday']]='Tuesday'
all_responses.loc[all_responses.weekday==3, ['weekday']]='Wednesday'
all_responses.loc[all_responses.weekday==4, ['weekday']]='Thursday'
all_responses.loc[all_responses.weekday==5, ['weekday']]='Friday'
all_responses.loc[all_responses.weekday==6, ['weekday']]='Saturday'
all_responses.loc[all_responses.weekday==0, ['weekday']]='Sunday'
```

```python
all_responses
```

```python
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
plt.figure(figsize=(15, 8))    
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)    
  
# Ensure that the axis ticks only show up on the bottom and left of the plot.    
# Ticks on the right and top of the plot are generally unnecessary chartjunk.    
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()
ax.set_title('Last Week Responses', fontsize= 20)
ax.set_ylabel('Responses', fontsize = 16)
ax.set_xlabel('Days', fontsize = 16)

plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
experts = all_responses['expert'].unique()
i=0
for e in experts:
    data= all_responses.loc[all_responses['expert'] == e]
    if(i!=1 and i!=3):
        data.plot(x ='weekday', y = 'responses', ax=ax, lw=2.5, rot=45, color=tableau20[i])
    i+=1
    
```
